```python
from flask import Flask, render_template, url_for, flash, redirect, request
from hash import *
import time,os
from IPython.display import display

app = Flask(__name__,template_folder='template')

app.config['UPLOAD_FOLDER'] = 'static'
app.config['SEND_FILE_MAX_AGE_DEFAULT'] = 0
n=20
app.config['TABLE']=empty_hashtable(n)


@app.route("/")
def home():
    return render_template("main_page.html")

def form():
    first_name = request.form.get("fname")
    last_name = request.form.get("lname")
    return first_name,last_name

def save_img(tbl_viz):
    if os.path.exists('static/bucket.png'):
        os.remove('static/bucket.png')
        os.remove('static/bucket')
    tbl_viz.render('static/bucket',format='png', view=False)
    time.sleep(1)
    bkt_img_path = os.path.join(app.config['UPLOAD_FOLDER'], 'bucket.png')
    return bkt_img_path

@app.route("/linear",methods=['GET','POST'])
def linear():
    if request.method == "POST":
        first_name, last_name = form()
        h=hash_function(first_name)
        table=app.config['TABLE']
        table=populate_hashtable_linear_probe((first_name,last_name),table)
        tbl_viz=objviz(table)
        bkt_img=save_img(tbl_viz)
        return render_template("table.html", user_image=bkt_img)
    return render_template("linear.html")


@app.route("/create",methods=['GET','POST'])
#@app.route("/chaining",methods=['GET','POST'])
def create():
    if request.method == "POST":
        first_name, last_name = form()
        h=hash_function(first_name)
        table=app.config['TABLE']
        table=populate_hashtable((first_name,last_name),table)
        tbl_viz=objviz(table)
        bkt_img=save_img(tbl_viz)
        return render_template("table.html", user_image=bkt_img)
    return render_template("form.html")


@app.route("/display",methods=['GET','POST'])
def display():
    table=app.config['TABLE']
    tbl_viz=objviz(table)
    bkt_img=save_img(tbl_viz)
    return render_template("t.html", user_image=bkt_img)    

app.run()
```
